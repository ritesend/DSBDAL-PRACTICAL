```python
# Calculating Covariance without using Pandas library
def covariance(x,y):
    mean_x = sum(x)/len(x)
    mean_y = sum(y)/len(y)
    sub_x = [i - mean_x for i in x]
    sub_y = [i - mean_y for i in y]
    numerator = sum([sub_x[i]*sub_y[i] for i in range(len(sub_x))])
    denominator = len(x)-1
    cov = numerator/denominator
    return cov
```

```python
for i in [0,1,2,3]:
    for j in [0,1,2,3]:
        if(i<j and i != j):
            val = covariance(df[features[i]],df[features[j]])
            print('Covariance for {} and {} : {}'.format(features[i],features[j],val))
    
```

```python
# Calculating Covariance using Pandas library
for i in [0,1,2,3]:
    for j in [0,1,2,3]:
        if(i<j and i != j):
            val = df[features[i]].cov(df[features[j]])
            print('Covariance for {} and {} : {}'.format(features[i],features[j],val))
    
```

```python
# Calculating Corelation without using Pandas library
def corelation(x,y):
    mean_x = sum(x)/float(len(x))
    mean_y = sum(y)/float(len(y))
    sub_x = [i - mean_x for i in x]
    sub_y = [i - mean_y for i in y]
    # Covariance of x and y
    numerator = sum([sub_x[i]*sub_y[i] for i in range(len(sub_x))])
    # standard deviation of x and y
    std_deviation_x = sum([sub_x[i] ** 2.0 for i in range(len(sub_x))])
    std_deviation_y = sum([sub_y[i] ** 2.0 for i in range(len(sub_y))])
    # Squaring by 0.5 to find the square root
    denominator = (std_deviation_x*std_deviation_y)**0.5
    cor = numerator/denominator
    return cor
```

```python
for i in [0,1,2,3]:
    for j in [0,1,2,3]:
        if(i<j and i != j):
            val = corelation(df[features[i]],df[features[j]])
            print('Corelation coefficient for {} and {} : {}'.format(features[i],features[j],val))   
```

```python
for i in [0,1,2,3]:
    for j in [0,1,2,3]:
        if(i<j and i != j):
            val = df[features[i]].corr(df[features[j]])
            print('Corelation coefficient for {} and {} : {}'.format(features[i],features[j],val))   
```
